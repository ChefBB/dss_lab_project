```python
import time
import random
import musicbrainzngs as mb

for artist in artists:
    query = dr.build_query(artist)
    result = None
    while not result:
        try:
            result = mb.search_artists(query)
            count = result.get('artist-count', 0)
            if count != 1:
                print(query, count)
        except Exception as e:
            print("Error:", e, "Query:", query)
    
        # wait 1–1.5 seconds to avoid rate limiting
        time.sleep(random.uniform(1, 1.5))
```
